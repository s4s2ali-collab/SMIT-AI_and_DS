### 43. Ek simple calculator banayein (+, -, *, /) symbols ke basis par operation perform karein.

```python
calc_1 = float(input("Enter first number: "))
calc_2 = float(input("Enter second number:"))
oprtr = input("Enter operator (+, -, /, *): ")

if oprtr == "+":
    result = calc_1 + calc_2
elif oprtr == "-":
    result = calc_1 - calc_2    
elif oprtr == "/":
    if calc_2 == 0:
        result = "Error! Division by Zero."
    else:
        result = calc_1 / calc_2
elif oprtr == "*":
    result = calc_1 * calc_2
else:
    result = "Enter valid operator."


print(f"Result: {result}")
```

### 44. User ka weight aur height le kar BMI category batayein (Underweight, Normal, Overweight).

```python
weight_kg = float(input("Enter weight in kg: "))
height_mt = float(input("Enter height in meters: "))

BMI = weight_kg / (height_mt ** 2)

if BMI < 18.5:
    category = "Underweight"
elif 18.5 <= BMI < 25:
    category = "Normal"
elif 25 <= BMI < 30:
    category = "Overweight"
else:
    category = "Obese"

print(f"Your BMI is: {BMI:.2f} and you are {category}")
```

### 47. Largest of Three: User se 3 numbers lein aur nested if-else use karke sabse bada number dhoondein.

```python
first_num = int(input("Enter first number: "))
second_num = int(input("Enter second number: "))
third_num = int(input("Enter third number: "))

if first_num >= second_num:
    if first_num >= third_num:
        largest = first_num
    else:
        largest = third_num
else:
    if second_num >= third_num:
        largest = second_num
    else:
        largest = third_num

print(f"The largest number is: {largest}")
```
